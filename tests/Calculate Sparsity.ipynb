```python
import h5py
import cloudpickle
import tensorflow as tf
import os
import sys
sys.path.append('..\\')
from Vocabulary import *
import re

def load_co_occurence(name):
    tmp_hf = h5py.File(name, "r")
    coocurrence = tmp_hf.get("co-ocurrence")[:]
    tf_co_occurences = tf.convert_to_tensor(coocurrence,dtype=tf.dtypes.float32)
    tmp_hf.close()
    return tf_co_occurences

vocab = Vocabulary()
vocab.load('..\\vocabs\\baseline')
length = vocab.get_size()
density = 0.0
dens_diag = 0.0
path = 'E:\\base_coocurrence_hdf_5000'
count = 0
for file_name in os.listdir(path):
    if 'diag' in file_name:
        continue
    try:
        c = load_co_occurence(path + '\\' + file_name)
    except:
        print(file_name,'error')
        continue
        
    
    zero = tf.constant(0, dtype=tf.float32)
    where = tf.not_equal(c, zero)
    indices = tf.where(where)
    
    regex_string = 'co_occurence_([0-9]{1,})_([0-9]{1,}).hdf5'
    x = re.match(regex_string,file_name)
    i = int(x.group(1))
    j = int(x.group(2))
    print(i,j)
    
    rest_length_x = min(5000,length - i * 5000)
    rest_length_y = min(5000,length - j * 5000)
    a= indices.shape[0] / (rest_length_x * rest_length_y)
    print(file_name, a)
    density += a
    dens_diag += (a + 5000)
    count += 1
    
avg_dens = density / count
print('sparsity'+ str(1 - avg_dens))
print('density' + str(avg_dens))

avg_dens = dens_diag / count
print('sparsity'+ str(1 - avg_dens))
print('density' + str(avg_dens))

    


    
```
