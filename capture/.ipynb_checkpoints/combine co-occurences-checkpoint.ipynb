```python
def save_dict(dictionary,i,j):
    path = 'S:\\tmp\\coocurrence_blocks'
    filepath = path + '\\block_{ni}_{nj}'.format(ni=i,nj=j)
    with open(filepath, 'wb') as file:
        cloudpickle.dump(dictionary, file)
    
```

```python
path = r'S:\\tmp\coocurence data'
import collections
file_list = os.listdir(path)
from Vocabulary import *
from scipy.sparse import dok_matrix,csr_matrix,lil_matrix
import scipy

vocab = Vocabulary()
vocab.load('../vocabs/baseline')
size = vocab.get_size()


block_size = 20000 # must be int
blocks_amount = math.ceil(vocab.get_size()/float(block_size))

#init
dict_of_dicts = {}
dict_list = []
for i in range(blocks_amount):
    for j in range(blocks_amount):
        dict_of_dicts[(i,j)]={}
        dict_list.append((i,j))
# track the state of the dict, if it is done, than saveset to None and remove the fucking dict
#dict_list = [(1,0),(2,1),(3,2),(4,3),(5,4),(6,5),(7,6)]
number_to_take = 8

while len(dict_list) > 0:
    current_dicts = dict_list[0:number_to_take]
    dict_list = dict_list[number_to_take:]
        
    for file_name in file_list:
        print(file_name)
        capturer = Co_Occurence_Capturer()
        capturer.load_co_occurence(path + "/" + file_name)
        coOcc_local = capturer.co_occurences
        print('length of in file (capturer.co_occurences) ' + str(len(coOcc_local)))
        for (x,y) in coOcc_local:
            if x < y:#co occurence is symetrical, zeilen präferiert => für asymetrisches feature
                continue
            dict_pos_x = int(x / block_size)
            dict_pos_y = int(y / block_size)
            position_tuple = (dict_pos_x,dict_pos_y)
            if(position_tuple in current_dicts):
                tuple_dict = dict_of_dicts[(dict_pos_x,dict_pos_y)]
                x_new = x % block_size
                y_new = y % block_size
                if (x_new,y_new) in tuple_dict:
                    tuple_dict[(x_new,y_new)] += coOcc_local[(x,y)]
                else:
                    tuple_dict[(x_new,y_new)]  = coOcc_local[(x,y)] #Apply the shifted log
    for (i,j) in current_dicts:
        print('saving i,j')
        tuple_dict = dict_of_dicts[(i,j)]
        save_dict(tuple_dict,i,j)
        #free memory
        tuple_dict = None
        dict_of_dicts[(i,j)] = None
```
