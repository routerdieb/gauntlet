```python
import collections
from Vocabulary import *
from scipy.sparse import dok_matrix,csr_matrix,lil_matrix
import scipy

vocab = Vocabulary()
vocab.load('../vocabs/wiki2021base')
size = vocab.get_size()


block_size = 20000 # must be int
blocks_amount = math.ceil(vocab.get_size()/float(block_size))

#init dicts
dict_of_dicts = {}
for i in range(blocks_amount):
    for j in range(blocks_amount):
        dict_of_dicts[(i,j)] = {}
        
take_before_write = 3

file_list = os.listdir(pathIn)
for file_index,file_name in enumerate(file_list):
    print(file_name)
    capturer = Co_Occurence_Capturer()
    capturer.load_co_occurence(pathIn + "/" + file_name)
    print('length of in file (capturer.co_occurences) ' + str(len(capturer.co_occurences)))
    
    for (x,y) in capturer.co_occurences:
        if symmetrie and x < y:#co occurence is symetrical, zeilen präferiert => für asymetrisches feature
            continue
        dict_pos_x = int(x / block_size)
        dict_pos_y = int(y / block_size)
        position_tuple = (dict_pos_x,dict_pos_y)
        tuple_dict = dict_of_dicts[(dict_pos_x,dict_pos_y)]
        x_new = x % block_size
        y_new = y % block_size
        if (x_new,y_new) in tuple_dict:
            tuple_dict[(x_new,y_new)] += capturer.co_occurences[(x,y)]
        else:
            tuple_dict[(x_new,y_new)]  = capturer.co_occurences[(x,y)]
    capturer = None #release memory
            
    if file_index > 0 and file_index % take_before_write == 0:
        print('saving i,j')
        for (i,j) in dict_of_dicts:
            tuple_dict = dict_of_dicts[(i,j)]
            if len(tuple_dict) > 0:#This is untested again.
                save_dict(pathOut,tuple_dict,i,j)
            dict_of_dicts[(i,j)] = {}
```
