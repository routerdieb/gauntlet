```python
import h5py
import sys
sys.path.append("..")
import os
from flair.models import MultiTagger
# Construction 2
from spacy.lang.en import English
nlp = English()
# Create a Tokenizer with the default settings for English
# including punctuation rules and exceptions
tokenizer = nlp.tokenizer
```

```python
from Vocabulary import *
from flair.tokenization import SegtokSentenceSplitter

vocab = Vocabulary()

path = 'S:\data'
dir_list = os.listdir(path)

for directory_name in dir_list:
    print(directory_name)
    for file_name in os.listdir(path + "/" + directory_name):
        print(file_name)
        filepath = path + "/" + directory_name + "/" + file_name
        filepath_new = path + "/" + directory_name + "/tokenised_" + file_name
        with open(filepath,'r',encoding='utf8') as read, open(filepath_new,'w',encoding='utf8') as writer:
            text = read.readlines()
            for line in text:
                if line.startswith('<doc'):
                    writer.write(line)
                elif line.startswith('</doc'):
                    writer.write(line)
                else:
                    tokens = tokenizer(line)
                    for token in tokens:
                        writer.write(token.text + " ")
```
